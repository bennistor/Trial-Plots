```python
# Display of ANZ region with a basemap

import numpy as np

from matplotlib.pyplot import figure, pcolor, colorbar, xlabel, ylabel, title, draw, quiver, show, savefig, axes
from cartopy import config, crs

figure1 = figure(figsize=(13, 10), dpi=80, facecolor='w', edgecolor='k')
rp = crs.RotatedPole(pole_longitude=141.38,pole_latitude=60.31, central_rotated_longitude=180.0)

ax = axes(projection=rp)
ax.coastlines('50m', linewidth=0.8)

show()
```
